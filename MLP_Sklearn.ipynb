```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score , confusion_matrix
dataset = load_digits()
x_train , x_test , y_train , y_test = train_test_split(dataset.data , dataset.target , test_size=0.2)
# we are creating am MLP here
NN=MLPClassifier(hidden_layer_sizes=(10) , max_iter=10000)
# for the fit we are training the neural network
NN.fit(x_train,y_train)
# for pred is basically predicting the values
y_pred = NN.predict(x_test)
# the accuracy score is decided here by comparing the test outputs with the predicted outputs
print(y_test )
print( y_pred )
print("these are the values")
score = accuracy_score(y_test , y_pred)*100
print(f"Accuracy for the neural network is {score}")

import matplotlib.pyplot as plt
plt.gray()
plt.matshow(dataset.images[1790])
plt.show()
print(dataset.data[1790])
```

we will be doing NLP next
there are multiple aspects here
- Lowercasing
- Tokenization
- Stopword Removal => are , the , they , prepositions etc
- Strmming => play , playing -- play
- Lemmatization => more accurate root words

```python
import nltk
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer , WordNetLemmatizer


text = "Cats are faster than Dogs"

text = text.lower()
tokens = word_tokenize(text)
#Stop words removal we need to remove prepositions and helping words which does not have a meaning alone
stop_words = set(stopwords.words("english"))
# we need to make an list of w from tokens which  are not there in the stop_words
# stop words are a set of words which are not supposed t
filtered = [w for w in tokens if w not in stop_words]

# print(filtered)


stemmer = PorterStemmer()
stemmed = [stemmer.stem(w) for w in filtered]

lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(w) for w in filtered]

print("Tokens" , tokens)
print("Without stopwords " , filtered)
print("Stemmed" , stemmed)
print("Lemmatized" , lemmatized)
```
